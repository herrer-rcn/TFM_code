```python
def organize_files_by_category(input_folder, output_folder):
    # Crear el directorio de salida si no existe
    os.makedirs(output_folder, exist_ok=True)

    # Obtener la lista de archivos en la carpeta de entrada
    files = [f for f in os.listdir(input_folder) if f.endswith('.csv')]

    for file in files:
        # Separar el nombre del archivo en partes
        parts = file.split('_')
        # Obtener el prefijo que incluye todo menos el último número
        category = '_'.join(parts[:-1])

        # Crear la carpeta de destino si no existe
        category_folder = os.path.join(output_folder, category)
        os.makedirs(category_folder, exist_ok=True)

        # Ruta completa del archivo de entrada y de salida
        input_path = os.path.join(input_folder, file)
        output_path = os.path.join(category_folder, file)

        # Copiar el archivo a la carpeta correspondiente
        shutil.copy(input_path, output_path)
        print(f"Archivo {file} copiado a {category_folder}")

# Directorios
input_folder = r"C:\Users\Holacons\OneDrive - Vueling Airlines, S.A\Escritorio\TFM\fragmentos"
output_folder = r"C:\Users\Holacons\OneDrive - Vueling Airlines, S.A\Escritorio\TFM\fragmentos_por_golpes_jugador"

# Organizar los archivos
organize_files_by_category(input_folder, output_folder)

```

```python

def get_time_ranges(input_folder):
    # Diccionario para almacenar los rangos de tiempo de cada archivo
    time_ranges = {}

    # Recorrer todas las carpetas y archivos dentro del directorio de entrada
    for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)
                df = pd.read_csv(file_path)
                start_time = df['TIMESTAMP'].iloc[0]
                end_time = df['TIMESTAMP'].iloc[-1]
                time_ranges[file] = (start_time, end_time)
                print(f"Archivo {file} procesado en {root}")

    return time_ranges

# Directorio de entrada
input_folder = r'C:/Users/Holacons/OneDrive - Vueling Airlines, S.A/Escritorio/TFM/fragmentos_por_golpes_jugador'

# Obtener los rangos de tiempo
time_ranges = get_time_ranges(input_folder)

# Mostrar los rangos de tiempo
for file, (start_time, end_time) in time_ranges.items():
    print(f"{file}: {start_time} - {end_time}")
```
