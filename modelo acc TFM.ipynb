```python
def leer_csvs_carpeta_en_dataframe(input_folder):
    # Lista para almacenar los DataFrames individuales
    dataframes = []
    
    # Iterar sobre cada archivo en la carpeta
    for file_name in os.listdir(input_folder):
        if file_name.endswith('.csv'):
            file_path = os.path.join(input_folder, file_name)
            # Leer el archivo CSV y aÃ±adirlo a la lista de DataFrames
            df = pd.read_csv(file_path)
            dataframes.append(df)
    
    # Concatenar todos los DataFrames en uno solo
    df_combinado = pd.concat(dataframes, ignore_index=True)
    
    return df_combinado

# Uso de la funciÃ³n
ruta_carpeta = "C:/Users/Holacons/OneDrive - Vueling Airlines, S.A/Escritorio/TFM/para_modelo"
df_combinado = leer_csvs_carpeta_en_dataframe(ruta_carpeta)

# Mostrar el DataFrame resultante
df_combinado.head()
```

```python
from sklearn.model_selection import GridSearchCV
# Separar caracterÃ­sticas (X) y el target (y)
X = df_final.drop(columns=['TIPO_GOLPE', 'JUGADOR'])
y = df_final['TIPO_GOLPE']
groups = df_final['JUGADOR']
# Definir el espacio de bÃºsqueda de hiperparÃ¡metros
param_grid = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size': [10, 20, 30],
    'p': [1, 2]  # Para la distancia de Minkowski
}
knn = KNeighborsClassifier(
    algorithm='auto',
    leaf_size=30,
    metric='minkowski',
    metric_params=None,
    n_jobs=-1,
    n_neighbors=5,
    p=2,
    weights='uniform'
)
group_kfold = GroupKFold(n_splits=6)
# Crear el objeto GridSearchCV
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=group_kfold, scoring='accuracy', n_jobs=-1)

# Ejecutar la bÃºsqueda en cuadrÃ­cula
grid_search.fit(X, y, groups=groups)

# Obtener los mejores hiperparÃ¡metros
best_params = grid_search.best_params_
print("Mejores hiperparÃ¡metros encontrados:")
print(best_params)

# Obtener la mejor puntuaciÃ³n
best_score = grid_search.best_score_
print("Mejor puntuaciÃ³n de validaciÃ³n cruzada:", best_score)
```
